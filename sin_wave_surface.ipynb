```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
%matplotlib inline
```

```python
theta1 = np.sin(np.linspace(0, 4*np.pi, 200))
theta2 = np.sin(np.linspace(0, 4*np.pi, 200))
T1, T2 = np.meshgrid(theta1, theta2)
Z = (T1+T2)/10

x = np.linspace(0, len(Z)*10, 200)
y = np.linspace(0, len(Z)*10, 200)
X, Y = np.meshgrid(x, y)
X.shape, Y.shape, Z.shape
```

```python
fig = plt.figure(dpi=150)
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface \
(X, Y, Z, cmap='copper', linewidth=1,
antialiased=True, alpha=1, rstride=3, cstride=3,)
# Labels.
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Elevation')

ax.set_zlim(-0.5, 0.5)

ax.elev = 45
ax.dist = 10
ax.azim = 148
```
